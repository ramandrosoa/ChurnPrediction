## Understanding the Datasets

### Train vs. Test
In this competition, you’ll gain access to two datasets that are samples of past subscriptions of a video streaming platform that contain information about the customer, the customers streaming preferences, and their activity in the subscription thus far. One dataset is titled `train.csv` and the other is titled `test.csv`.

```python
from scipy.stats import ttest_ind
```

```python
    #Bivariate Analysis - NUMERICAL FEATURES
    #Statistical hypothesis to compare the means of the 2 groups (check if they are significantly different)
significance_level = 0.05
drop_columns = []

for col in numerical_cols :
        churn_yes = train_df[train_df['Churn'] ==1][col]
        churn_no = train_df[train_df['Churn']==0][col]
        t_stat, p_value = ttest_ind(churn_yes, churn_no)
        print(f'T-test for {col}: t-stat = {t_stat:.3f}, p-value = {p_value}:.3f')
        if p_value > significance_level:
                drop_columns.append(col)
drop_columns
```

By comparing the mean of the 2 groups in each feature, a high t-statistic (significantly different from 0) indicates that the means are significantly different. A low p-value suggests that the feature may be valuable for the model

```python
    #Split the data
X_train = train_df.copy()
y_train = X_train.pop("Churn")
```

```python
def preprocessing(X) :
    
        # Drop specified columns
    X = X.drop(columns=['TotalCharges', 'PaperlessBilling', 
                        'MultiDeviceAccess', 'DeviceRegistered', 
                        'CustomerID'])
    
    # Identify numerical and categorical columns
    num_cols = X.select_dtypes(include=['number']).columns
    cat_cols = X.select_dtypes(include=['object']).columns
    
    # Scaling and encoding
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), num_cols),
            ('cat', OneHotEncoder(drop="first", sparse=False), cat_cols)
        ])
    
    # Fit and transform X    
    X_processed = preprocessor.fit_transform(X)
    
    # Create DataFrame with processed data
    num_features = num_cols.tolist()
    
    # Get the feature names for the OneHotEncoded columns
    cat_features = preprocessor.named_transformers_['cat'].get_feature_names(cat_cols)
    features = num_features + cat_features.tolist()
    
    X_df = pd.DataFrame(X_processed, columns=features)
    
    #Normalization 
    normalizer = MinMaxScaler()
    X_norm = normalizer.fit_transform(X_df)
    X_norm_df = pd.DataFrame(X_norm, columns = features)
    
    return X_norm_df
```

```python
X_train_norm = preprocessing(X_train)
X_train_norm.tail()
```
